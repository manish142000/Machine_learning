```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data_cleaned.csv')
data.head()
```

```python
#segregating dependent and independent variables
x = data.drop(['Survived'], axis = 1)
y = data['Survived']
x.shape, y.shape
```

```python
#splitting the data into train and test 
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, random_state = 56)
```

```python
#normalising using min_max scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
cols = train_x.columns
cols
```

```python
train_x_scaled = scaler.fit_transform(train_x)
train_x =  pd.DataFrame(train_x_scaled, columns = cols)
train_x.head()
```

```python
test_x_scaled = scaler.transform(test_x)
test_x = pd.DataFrame(test_x_scaled, columns = cols)
test_x.head()
```

```python
from sklearn.linear_model import LogisticRegression as LogReg
from sklearn.metrics import f1_score
```

```python
logreg = LogReg()

logreg.fit(train_x, train_y)
```

```python
train_predict = logreg.predict(train_x)
train_predict
```

```python
k = f1_score(train_predict, train_y)
k
```

```python
test_predict = logreg.predict(test_x)
test_predict
```

```python
k = f1_score(test_predict, test_y)
k
```

Making predictions using predict_probability function

```python
train_predict = logreg.predict_proba(train_x)
train_predict
```

```python
train_preds = train_predict[:, 1]
train_preds
```

```python
for i in range(0, len(train_preds)):
    if train_preds[i] >= 0.5:
        train_preds[i] = 1
    else:
        train_preds[i] = 0
```

```python
k = f1_score(train_preds, train_y)
k
```

Confusion matrix

```python
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(test_predict, test_y)
cf
```

```python
from sklearn.metrics import classification_report as rep
print(rep(test_predict, test_y))
```

```python
logreg.coef_
```

```python
plt.figure(figsize=(8, 6), dpi=120, facecolor='w', edgecolor='b')
x = range(len(train_x.columns))
c = logreg.coef_.reshape(-1)
plt.bar( x, c )
plt.xlabel( "Variables")
plt.ylabel('Coefficients')
plt.title('Coefficient plot')
```

```python
Coefficients = pd.DataFrame(
        {
            'Variable' : train_x.columns,
            'Coeffiecients': abs(c)
        }
)
```

```python
Coefficients.head()
```

```python
significant_variables = Coefficients.loc[Coefficients.Coeffiecients > 0.3]
```

```python
subset = data.loc[:][significant_variables['Variable'].values]
```

```python
subset
```

```python
train_x, test_x, train_y, test_y = train_test_split(subset, y, random_state = 56)
```

```python
logreg = LogReg()

logreg.fit(train_x, train_y)
```

```python
train_predict = logreg.predict(train_x)
k = f1_score(train_predict, train_y)
print('training_f1_score', k)
```
