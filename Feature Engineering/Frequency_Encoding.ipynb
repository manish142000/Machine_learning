```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train_bm.csv')
df.head()
```

```python
Item_Type_Count = df.Item_Type.value_counts(normalize = True)
Item_Type_Count
```

```python
df['Item_Type_Count'] = df.Item_Type.apply(lambda x: Item_Type_Count[x])
df.head()
```

Taking the mean of frequency of Item_Type

```python
Mean_Item_Type_Count = df.groupby('Item_Type')['Item_Outlet_Sales'].mean()
df['Mean_Item_Type_Count'] = df.Item_Type.apply(lambda x: Mean_Item_Type_Count[x])
df.head()
```
