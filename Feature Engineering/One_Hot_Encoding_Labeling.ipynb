Solutions For the above problems

Problem 1

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.Outlet_Size.value_counts()
```

```python
le.fit_transform(['small', 'medium', 'high'])
```

Since Label encoder uses alphabetical order to assign values we cannot use LabelEncoder. Instead We use Map function

```python
df.Outlet_Size = df.Outlet_Size.map( { 'Small' : 0, 'Medium': 1, 'High': 2 } )
```

```python
df.Outlet_Size.head()
```

Problem 2

```python
df.Item_Identifier.value_counts()
```

We can see that there are Items with identifiers less than 2.

```python
Item_Identifier_values = df.Item_Identifier
Item_Identifier_values
```

```python
counts = df.Item_Identifier.value_counts()
counts
```

```python
for index, item in enumerate(Item_Identifier_values):
    if counts[item] < 4:
        Item_Identifier_values[index] = "other"
```

```python
Item_Identifier_values.value_counts()
```

```python
df.Item_Identifier = Item_Identifier_values
df.head(7)
```
