```python
import pandas as pd
import numpy as np
df = pd.read_csv('../input/date-time-handling/messages.csv')
df.head()
```

```python
df.date = pd.to_datetime(df.date)
df.dtypes
```

```python
df.date.dt.hour.head()
```

```python
df.date.dt.minute.head()
```

```python
additional_df = pd.DataFrame(
{
    "year": df.date.dt.year,
    "month": df.date.dt.month,
    "hour": df.date.dt.hour,
    "day": df.date.dt.day,
    "dayofyear": df.date.dt.dayofyear,
    "week": df.date.dt.week,
    "dayofweek": df.date.dt.dayofweek,
    "quarter": df.date.dt.quarter,
}
)
additional_df.head()
```

```python
df = pd.concat([df, additional_df], axis=1)
df.head()
```

Checking if the date falls on the weekday or the weekend

```python
df['is_weekday']=0 

for i in range(0, len(df)):
    if ((df['dayofweek'][i] == 5) | (df['dayofweek'][i] == 6)):
        df['is_weekday'][i] = 0
    else: 
        df['is_weekday'][i] = 1
```
